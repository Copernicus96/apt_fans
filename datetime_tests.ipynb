```python
apt_temps = pd.DataFrame()
for i, chx in enumerate([ch1, ch2, ch3, ch4, ch5]):
    apt_temps['temp_ch{}'.format(i+1)] = chx['Temperature(F)']
```

```python
dtypes_apt = {'Time': 'str', 'Item': 'str', 'State': 'float'}
items = pd.read_csv('apt_items.csv', index_col = 0, parse_dates = [0], dtype = dtypes_apt)
```

```python
items 
```

```python
lower = items.loc[items['Item'] == 'Lower Fan']
upper = items.loc[items['Item'] == 'Upper Fan']
thermo = items.loc[items['Item'] == 'Thermostat']
```

```python
# add data about fan status in new columns to each temp dataset
for fan in [lower, upper]:
    for index, row in fan.iterrows():
        apt_temps.loc[index:, row['Item']] = row['State']
# add data about thermostat temp        
for index, row in thermo.iterrows():
        apt_temps.loc[index:, row['Item']] = row['State']
```

```python
apt_temps
```

```python
temp_diffs = pd.DataFrame()
for i in range(1,6):
    temp_diffs['diff_ch{}'.format(i)] = apt_temps['temp_ch{}'.format(i)] - apt_temps['Thermostat'] 

temp_diffs['Lower Fan'] = apt_temps['Lower Fan']
temp_diffs['Upper Fan'] = apt_temps['Upper Fan']
```

```python
both_fan = (temp_diffs['Lower Fan'] == 1) & (temp_diffs['Upper Fan'] == 1)
lower_fan = (temp_diffs['Lower Fan'] == 1) & (temp_diffs['Upper Fan'] == 0)
upper_fan = (temp_diffs['Lower Fan'] == 0) & (temp_diffs['Upper Fan'] == 1)
no_fan = (temp_diffs['Lower Fan'] == 0) & (temp_diffs['Upper Fan'] == 0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
lbls = ['Both', 'Lower Only', 'Upper Only', 'Neither']
for i, subset in enumerate([both_fan, lower_fan, upper_fan, no_fan]):
    sub = temp_diffs[subset]
    
    # Draw the density plot
    sns.distplot(sub['diff_ch3'], hist = False, kde = True, kde_kws = {'linewidth': 3}, label = lbls[i])
    
# Plot formatting
plt.legend(prop={'size': 12}, title = 'Fans Running')
plt.title('Apartment Loft Temperature Deviation')
plt.xlabel('Temperature Difference from Thermostat')
plt.ylabel('Density')
```
